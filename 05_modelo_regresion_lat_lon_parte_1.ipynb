```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score

from sklearn.preprocessing import MinMaxScaler

from geopy.geocoders import Nominatim
```

### Archivo CSV

```python
dir_procesado = 'data/procesado/'
archivo = 'datos_crudos.csv'
df = pd.read_csv(f'{dir_procesado}{archivo}')

df.head(2)
```

### Columna "vendor"

```python
df_col = df.groupby('vendor').agg({'id':'count'})
df_col.reset_index(drop=False, inplace=True)
df_col.sort_values(['id'], ascending=False, inplace=True)

total_col = df_col['id'].sum()
df_col['total_porcentaje'] = df_col['id'].apply(lambda x: x/total_col*100)
df_col['total_porcentaje_acumulado'] = df_col['total_porcentaje'].cumsum()
```

```python
df_col.head(10)
```

```python
lista_col = df_col['vendor'].values[:5]
lista_col
```

### Conjunto de datos para modelo

```python
df_modelo = df.copy()
df_modelo['vendor_mod'] = df_modelo['vendor'].apply(lambda x: x if x in lista_col else 'otros')
```

```python
df_modelo.info()
```

### One Hot Encoder

```python
df_ohe = pd.get_dummies(df_modelo, columns=['vendor_mod'])
df_ohe.head(3)
```

### Conjunto de datos para modelo X^2

```python
lista_x2 = [
    'days_on_site',
    'bathrooms',
    'parking_lots',
    'num_bedrooms',
    'm2']
```

```python
df_sq = df_ohe.copy()
for col in lista_x2:
    df_sq[f'{col}_x2'] = df_sq[col].apply(lambda x: x**2)
```

```python
df_sq.head(3)
```

### Conjunto de datos escalado

```python
df_sc = df_sq.copy()

lista_escala = [
    'days_on_site',
    'bathrooms',
    'parking_lots',
    'num_bedrooms',
    'm2',
    'days_on_site_x2',
    'bathrooms_x2',
    'parking_lots_x2',
    'num_bedrooms_x2',
    'm2_x2']

for col in lista_escala:
    scaler = MinMaxScaler()
    df_sc[f'{col}_sc'] = scaler.fit_transform(df_sc[[col]])
```

```python
df_sc.head(3)
```

### Dirección Lat-Lon

```python
geolocator = Nominatim(user_agent="reto_dd3")
```

```python
df_address = df_sc.copy()
```

```python
df_address['ciudad'] = df_address[['lat', 'lon']]\
    .apply(lambda x: geolocator.reverse((x[0], x[1])).raw['address']['city'], axis=1)
```

```python
# def get_colonia(lat, lon):
#     colonia = np.nan
#     try:
#         colonia = geolocator.reverse((lat, lon)).raw['address']['neighbourhood']
#     except:
#         colonia = np.nan
#     return colonia


# df_address['colonia'] = df_address[['lat', 'lon']].apply(lambda x: get_colonia(x[0], x[1]), axis=1)

# df_address['colonia_mod'] = df_address[['colonia', 'ciudad']].apply(lambda x: x[0] if str(x[0])!='nan' else x[1], axis=1)
```

```python
def get_codigo_postal(lat, lon):
    valor = np.nan
    try:
        valor = geolocator.reverse((lat, lon)).raw['address']['postcode']
    except:
        valor = np.nan
    return valor


df_address['codigo_postal'] = df_address[['lat', 'lon']].apply(lambda x: get_codigo_postal(x[0], x[1]), axis=1)
```

```python
df_address.head(3)
```

### Dirección One Hot Encoder

```python
df_address_cp = df_address.copy()
```

```python
df_temp_cp = df_address_cp.groupby('codigo_postal').agg({'id':'count'})
df_temp_cp.reset_index(drop=False, inplace=True)
df_temp_cp.sort_values(['id'], ascending=False, inplace=True)

total_col = df_temp_cp['id'].sum()
df_temp_cp['total_porcentaje'] = df_temp_cp['id'].apply(lambda x: x/total_col*100)
df_temp_cp['total_porcentaje_acumulado'] = df_temp_cp['total_porcentaje'].cumsum()
```

```python
df_temp_cp.head(3)
```

```python
lista_cp = df_temp_cp['codigo_postal'].values[:2]
lista_cp
```

```python
df_address_cp['codigo_postal_mod'] = df_address_cp['codigo_postal'].apply(lambda x: x if x in lista_cp else 'otro_cp')
```

```python
df_address_cp.head(2)
```

```python
df_address_cp_ohe = pd.get_dummies(df_address_cp, columns=['codigo_postal_mod'])
df_address_cp_ohe.head(3)
```

### Almacenamiento

```python
df_almacenamiento = df_address_cp_ohe.copy()
dir_almacenamiento = 'data/procesado/'
archivo_almacenamiento = 'datos_latitud_longitud.csv'
df_almacenamiento.to_csv(f'{dir_almacenamiento}{archivo_almacenamiento}', index=False)
print(archivo_almacenamiento)
```
